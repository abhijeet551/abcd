```python
import numpy as np
import matplotlib.pyplot as plt
import pyroomacoustics as pra
fs=1000000
t=1
f=40000
a=np.linspace(0,t,fs)
signal=np.sin(2*3.14*f*a) #signal wave
room=pra.ShoeBox([1.0,1.0,1.0],absorption=1.0,fs=fs) 
a=np.linspace(0.2,0.304,8) #random points spaced 13mm
b=list(a)
c=b.pop() # no.in  bracket is the position  of  transducer among 8 transducers we want to act as recceiver 
angle = np.array(45) #angle we want our beam at
d= 13*np.sin(np.deg2rad(angle))/(8*40000) #delay due to phase diff
for i in range(len(b)):
    room.add_source([b[i],0.2,0.2],delay=((b[i]-0.2)/13)*d)

room.plot()
plt.show()
```
